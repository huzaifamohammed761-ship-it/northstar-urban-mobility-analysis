```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the datasets from Github

```python
orders_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/orders.csv'

deliveries_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/deliveries.csv'

complaints_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/complaints.csv'

customers_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/customers.csv'

drivers_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/drivers.csv'

vehicles_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/vehicles.csv'

incidents_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/incidents.csv'

hubs_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/hubs.csv'

app_events_url = 'https://raw.githubusercontent.com/huzaifamohammed761-ship-it/northstar-urban-mobility-analysis/main/northstar_dataset/northstar_dataset/app_events.csv'


orders = pd.read_csv(orders_url)
deliveries = pd.read_csv(deliveries_url)
complaints = pd.read_csv(complaints_url)
customers = pd.read_csv(customers_url)
drivers = pd.read_csv(drivers_url)
vehicles = pd.read_csv(vehicles_url)
incidents = pd.read_csv(incidents_url)
hubs = pd.read_csv(hubs_url)
app_events = pd.read_csv(app_events_url)
```

Inspecting the data

```python
#Inspecting

dataframes = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "drivers": drivers,
    "vehicles": vehicles,
    "complaints": complaints,
    "incidents": incidents,
    "app_events": app_events,
    "hubs": hubs
}

for dataset_name, dataset_df in dataframes.items():
    print(f"\n{'-' * 60}")
    print(f"Inspecting Dataset: {dataset_name}")
    print(f"{'-' * 60}")

    print("\nDataset Dimensions:")
    print(dataset_df.shape)

    print("\nAvailable Columns:")
    print(list(dataset_df.columns))

    print("\nNull Value Count:")
    print(dataset_df.isna().sum())

    print("\nColumn Data Types:")
    print(dataset_df.dtypes)

    print("\nPreview of Dataset:")
    display(dataset_df.head())
```

Creating clean copies of the dataset

```python
# CREATE CLEAN COPIES OF DATASETS

customers_cleaned = customers.copy()
orders_cleaned = orders.copy()
deliveries_cleaned = deliveries.copy()
drivers_cleaned = drivers.copy()
vehicles_cleaned = vehicles.copy()
complaints_cleaned = complaints.copy()
incidents_cleaned = incidents.copy()
app_events_cleaned = app_events.copy()
hubs_cleaned = hubs.copy()
```

Standardizing the zone values

```python
# Standardizing Zone-Related Values

def normalize_zone_names(zone_column):

    standardized_zones = (
        zone_column.astype(str)
        .str.strip()
        .str.lower()
        .replace({
            'north': 'North',
            'south': 'South',
            'east': 'East',
            'west': 'West',
            'central': 'Central',
            'ctr': 'Central',
            'airport': 'Airport',
            'riverside': 'Riverside'
        })
        .str.title()
    )

    return standardized_zones


# Applying Zone Standardization

customers_cleaned['home_zone'] = normalize_zone_names(
    customers_cleaned['home_zone']
)

orders_cleaned['pickup_zone'] = normalize_zone_names(
    orders_cleaned['pickup_zone']
)

orders_cleaned['dropoff_zone'] = normalize_zone_names(
    orders_cleaned['dropoff_zone']
)

drivers_cleaned['base_zone'] = normalize_zone_names(
    drivers_cleaned['base_zone']
)

vehicles_cleaned['assigned_zone'] = normalize_zone_names(
    vehicles_cleaned['assigned_zone']
)

app_events_cleaned['zone_context'] = normalize_zone_names(
    app_events_cleaned['zone_context']
)

hubs_cleaned['zone'] = normalize_zone_names(
    hubs_cleaned['zone']
)

print("Zone values standardized successfully.")
```

Converting the date and time columns

```python
# Converting Date and Time Columns

customers_cleaned['signup_date'] = pd.to_datetime(
    customers_cleaned['signup_date'],
    errors='coerce'
)

orders_cleaned['order_created_at'] = pd.to_datetime(
    orders_cleaned['order_created_at'],
    errors='coerce'
)

deliveries_cleaned['dispatch_time'] = pd.to_datetime(
    deliveries_cleaned['dispatch_time'],
    errors='coerce'
)

deliveries_cleaned['delivery_completed_at'] = pd.to_datetime(
    deliveries_cleaned['delivery_completed_at'],
    errors='coerce'
)

complaints_cleaned['created_at'] = pd.to_datetime(
    complaints_cleaned['created_at'],
    errors='coerce'
)

incidents_cleaned['reported_at'] = pd.to_datetime(
    incidents_cleaned['reported_at'],
    errors='coerce'
)

app_events_cleaned['event_timestamp'] = pd.to_datetime(
    app_events_cleaned['event_timestamp'],
    errors='coerce'
)

print("Date columns converted successfully.")
```

Handling missing values

```python
# Handling Missing Values

customers_cleaned['loyalty_score'] = customers_cleaned[
    'loyalty_score'
].fillna(
    customers_cleaned['loyalty_score'].median()
)

customers_cleaned['preferred_channel'] = customers_cleaned[
    'preferred_channel'
].fillna(
    'Not Specified'
)


deliveries_cleaned['customer_rating_post_delivery'] = deliveries_cleaned[
    'customer_rating_post_delivery'
].fillna(
    deliveries_cleaned[
        'customer_rating_post_delivery'
    ].median()
)


vehicles_cleaned['battery_health_pct'] = vehicles_cleaned[
    'battery_health_pct'
].fillna(
    vehicles_cleaned['battery_health_pct'].median()
)

print("Missing values handled successfully.")
```

Feature Engineering and Operations


```python
# Creating Analytical Features

deliveries_cleaned['delivery_time_hours'] = (
    deliveries_cleaned['delivery_completed_at']
    - deliveries_cleaned['dispatch_time']
).dt.total_seconds() / 3600


deliveries_cleaned['risk_delivery_flag'] = (
    (deliveries_cleaned['manual_route_override_count'] >= 2) |
    (deliveries_cleaned['proof_of_completion_missing'] == 1)
).astype(int)


customer_complaint_totals = complaints_cleaned.groupby(
    'customer_id'
).size()


customers_cleaned['total_complaints'] = customers_cleaned[
    'customer_id'
].map(
    customer_complaint_totals
).fillna(0)

print("Feature engineering completed successfully.")
```

Integrated Operational Data View

```python
# Combining Operational Datasets

northstar_master_data = deliveries_cleaned.merge(
    orders_cleaned,
    on='order_id',
    how='left'
)

northstar_master_data = northstar_master_data.merge(
    customers_cleaned,
    on='customer_id',
    how='left'
)

northstar_master_data = northstar_master_data.merge(
    drivers_cleaned,
    on='driver_id',
    how='left'
)

northstar_master_data = northstar_master_data.merge(
    vehicles_cleaned,
    on='vehicle_id',
    how='left'
)

print("Final Integrated Dataset Shape:")
print(northstar_master_data.shape)

display(northstar_master_data.head())
```

Delivery Performance Analysis

```python
# Delivery Performance Analysis

delivery_performance_summary = northstar_master_data.groupby(
    'delivery_status'
).agg({
    'delivery_time_hours': 'mean',
    'manual_route_override_count': 'mean',
    'fuel_or_charge_cost': 'mean',
    'customer_rating_post_delivery': 'mean'
}).round(2)

display(delivery_performance_summary)
```

```python
# Correcting inconsistent delivery status logic

mask = (
    deliveries_cleaned['delivery_completed_at'].notna()
) & (
    deliveries_cleaned['delivery_status'] == 'Failed'
)

deliveries_cleaned.loc[
    mask,
    'delivery_status'
] = 'Completed'
```

```python
# Delivery Status Distribution

delivery_status_summary = northstar_master_data[
    'delivery_status'
].value_counts()

print(delivery_status_summary)
```

Manual Route Override Analysis

```python
# Average Manual Route Overrides by Delivery Status

route_override_summary = northstar_master_data.groupby(
    'delivery_status'
)['manual_route_override_count'].mean().round(2)

print(route_override_summary)
```

```python
# Visualizing Average Route Overrides by Delivery Status

plt.figure(figsize=(8, 5))

plt.plot(
    route_override_summary.index,
    route_override_summary.values,
    marker='o',
    linewidth=2,
    color='darkorange'
)

plt.title(
    'Average Manual Route Overrides by Delivery Status',
    fontsize=14
)

plt.xlabel('Delivery Status')
plt.ylabel('Average Override Count')

plt.grid(
    linestyle='--',
    alpha=0.4
)

plt.show()
```

Complaint and Delivery Failure Analysis

```python
# Complaint & Delivery Performance Analysis

complaint_analysis_data = complaints_cleaned.merge(
    orders_cleaned[['order_id', 'customer_id']],
    on='order_id',
    how='left'
)

complaint_analysis_data = complaint_analysis_data.merge(
    deliveries_cleaned[['order_id', 'delivery_status']],
    on='order_id',
    how='left'
)

complaint_performance_summary = complaint_analysis_data.groupby(
    'delivery_status'
).agg({
    'complaint_id': 'count',
    'compensation_amount': 'mean',
    'resolution_days': 'mean'
}).round(2)

display(complaint_performance_summary)
```

Zone Operational Analysis

```python
# Pickup Zone Operational Analysis

zone_performance_summary = northstar_master_data.groupby(
    'pickup_zone'
).agg({
    'delivery_time_hours': 'mean',
    'fuel_or_charge_cost': 'mean',
    'delivery_id': 'count',
    'customer_rating_post_delivery': 'mean'
}).round(2)

zone_performance_summary = zone_performance_summary.rename(columns={
    'delivery_id': 'total_deliveries',
    'delivery_time_hours': 'avg_delivery_time'
})

zone_performance_summary = zone_performance_summary.sort_values(
    by='avg_delivery_time',
    ascending=False
)

display(zone_performance_summary)
```

```python
# Delivery Status Distribution Across Pickup Zones

zone_delivery_status = northstar_master_data.pivot_table(
    index='pickup_zone',
    columns='delivery_status',
    values='delivery_id',
    aggfunc='count',
    fill_value=0
)

display(zone_delivery_status)
```

Average Delivery time visualization

```python
# Visualizing Average Delivery Time by Pickup Zone

plt.figure(figsize=(10, 6))

zone_performance_summary['avg_delivery_time'].plot(
    kind='bar',
    color='mediumseagreen',
    edgecolor='black'
)

plt.title('Average Delivery Time Across Pickup Zones')
plt.xlabel('Pickup Zone')
plt.ylabel('Average Delivery Time (Hours)')
plt.xticks(rotation=45)

plt.show()
```

DRIVER PERFORMANCE PROFILING

```python
# Driver Performance Evaluation

driver_performance_summary = northstar_master_data.groupby(
    'driver_id'
).agg({
    'delivery_id': 'count',
    'delivery_time_hours': 'mean',
    'manual_route_override_count': 'mean',
    'customer_rating_post_delivery': 'mean',
    'fuel_or_charge_cost': 'mean'
}).round(2)

driver_performance_summary = driver_performance_summary.rename(columns={
    'delivery_id': 'completed_deliveries',
    'delivery_time_hours': 'avg_delivery_time',
    'customer_rating_post_delivery': 'avg_customer_rating'
})

lowest_rated_drivers = driver_performance_summary.sort_values(
    by='avg_customer_rating'
).head(10)

display(lowest_rated_drivers)
```

```python
# Identifying High Risk Drivers

driver_performance_summary['risk_driver_flag'] = (
    (driver_performance_summary['manual_route_override_count'] > 1.2) |
    (driver_performance_summary['avg_customer_rating'] < 3.5)
)

risk_driver_distribution = driver_performance_summary[
    'risk_driver_flag'
].value_counts()

print(risk_driver_distribution)
```

Maintenance Analysis

```python
# Vehicle Maintenance and Operational Analysis

vehicle_performance_summary = northstar_master_data.groupby(
    'vehicle_type'
).agg({
    'delivery_time_hours': 'mean',
    'fuel_or_charge_cost': 'mean',
    'battery_health_pct': 'mean',
    'manual_route_override_count': 'mean'
}).round(2)

vehicle_performance_summary = vehicle_performance_summary.rename(columns={
    'delivery_time_hours': 'avg_delivery_time',
    'battery_health_pct': 'avg_battery_health',
    'manual_route_override_count': 'avg_route_overrides'
})

display(vehicle_performance_summary)
```

```python
# Maintenance Status Performance Analysis

maintenance_performance_summary = northstar_master_data.groupby(
    'maintenance_status'
).agg({
    'delivery_time_hours': 'mean',
    'customer_rating_post_delivery': 'mean',
    'fuel_or_charge_cost': 'mean'
}).round(2)

maintenance_performance_summary = maintenance_performance_summary.rename(columns={
    'delivery_time_hours': 'average_delivery_time',
    'customer_rating_post_delivery': 'average_customer_rating'
})

display(maintenance_performance_summary)
```

Mobile Platform and Application Analysis

```python
# Mobile Application Event Performance Analysis

platform_event_summary = app_events_cleaned.groupby(
    'event_type'
).agg({
    'api_latency_ms': 'mean',
    'success_flag': 'mean',
    'event_id': 'count'
}).round(2)

platform_event_summary = platform_event_summary.rename(columns={
    'api_latency_ms': 'average_api_latency',
    'success_flag': 'success_rate',
    'event_id': 'total_events'
})

display(platform_event_summary)
```

Correlation and Relationship Analysis
